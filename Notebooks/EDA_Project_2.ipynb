## **Statistical Analysis: Graphical Rep.**

**1. What is the Diagnosis Distribution of the Tumor Types: (M=Malignant, B=Benign)?**

```python
# Count values of diagnosis
diagnosis_counts = BreastCancer_data['diagnosis'].value_counts()

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(diagnosis_counts, labels=diagnosis_counts.index, autopct='%1.1f%%', startangle=90, colors=['Magenta', 'lightblue'], textprops={'fontsize': 12, 'weight': 'bold'}, wedgeprops={'linewidth': 2, 'edgecolor': 'black'})
plt.title('Diagnosis Distribution of the Tumor Types: (M=Malignant, B=Benign)', weight='bold', color='black')
plt.axis('equal')  # Perfect circular pie output
plt.show()
```

**Observations:**
*  From the diagnosis distribution pie chart, it can be inferred that the dataset is imbalanced, with a significantly higher proportion of benign tumors (62.7%) compared to malignant tumors (37.3%).

**2. What can you deduce from the descriptive statistics (mean, standard deviation, and skewness) for the 'mean' features in the breast cancer dataset?**

```python
# Select the columns representing the mean values of various breast cancer characteristics.
mean_features = [col for col in BreastCancer_data.columns if '_mean' in col]

# Calculate the mean, standard deviation, and skewness for the selected 'mean' features
stats = BreastCancer_data.groupby('diagnosis')[mean_features].agg(['mean', 'std', 'skew'])

# Display the calculated statistics, transpose it and with a background color gradient and grid
display(stats.T.style.background_gradient(cmap='inferno').set_properties(**{'border': '1px solid white'}).set_caption("Descriptive Statistics for Mean Features by Diagnosis"))
```

**Observations:**

*   On average, malignant tumors (M) exhibit higher mean values across most features (e.g., radius, perimeter, area) compared to benign tumors (B), indicating larger cell and tumor sizes.
*   The standard deviations for malignant tumors are generally larger than for benign tumors, suggesting greater variability in the characteristics of malignant cells.
*   Many features show different skewness values between benign and malignant diagnoses, implying that the distributions of these characteristics differ between the two tumor types.

**3. What can you deduce from the descriptive statistics (mean, standard deviation, and skewness) for the 'worst' features in the breast cancer dataset?**

```python
# Select the columns representing the worst values of various breast cancer characteristics.
worst_features = [col for col in BreastCancer_data.columns if '_worst' in col]

# Calculate the mean, standard deviation, and skewness for the selected 'worst' features.
stats = BreastCancer_data.groupby('diagnosis')[worst_features].agg(['mean', 'std', 'skew'])

# Display the calculated statistics, transpose it and include a background color gradient and grid.
display(stats.T.style.background_gradient(cmap='viridis').set_properties(**{'border': '1px solid white'}).set_caption("Descriptive Statistics for Worst Features by Diagnosis"))
```

**Observations from Descriptive Statistics of Worst Features:**

*   Similar to the mean features, malignant tumors (M) generally show higher mean values for the 'worst' features (e.g., radius_worst, perimeter_worst, area_worst) compared to benign tumors (B), reinforcing that malignant cells tend to have more extreme characteristics.
*   The standard deviations for the 'worst' features are also typically higher for malignant tumors, indicating greater variability in the most extreme measurements of malignant cells.
*   The skewness values for many 'worst' features differ between benign and malignant diagnoses, suggesting distinct distributional shapes for the extreme characteristics of the two tumor types.

**4. How do the distributions of the top 6 mean features differ between benign and malignant tumors?**

```python
# Select the top 6 mean features to visualize, based on initial analysis (you can adjust this list)
top_6_mean_features = ['radius_mean', 'perimeter_mean', 'area_mean', 'texture_mean', 'compactness_mean', 'concavity_mean']

# Create a figure to hold the violin plots
plt.figure(figsize=(15, 8))

# Loop through each selected feature and create a violin plot
for i, feature_name in enumerate(top_6_mean_features, 1):
    # Create a subplot for the current feature
    plt.subplot(2, 3, i)

    # Create a violin plot comparing the distribution of the feature for each diagnosis
    sns.violinplot(x='diagnosis', y=feature_name, data=BreastCancer_data, palette={'M':'Lightblue','B':'Magenta'}, split=True)

    # Set the title for the current subplot
    plt.title(f'The Distribution of {feature_name}')

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Display the plots
plt.show()
```

**Observations:**

*   For features like radius_mean, perimeter_mean, and area_mean, the violin plots for malignant tumors (M) are generally wider and shifted towards higher values compared to benign tumors (B), indicating that malignant tumors tend to be larger.
*   The violin plots for concavity_mean show a clearer separation between the two diagnosis groups, with malignant tumors exhibiting a wider range and higher density at larger concavity values than benign tumors.
*   The shapes of the violin plots reveal differences in the distribution of features beyond just the mean and standard deviation, suggesting that some features may have multi-modal distributions or different spread patterns between benign and malignant cases.

**5.  How do the distributions of various breast cancer features (such as radius, texture, perimeter, etc.) differ between benign and malignant tumors?**

```python
# List of Selected features to plot
selected_features = [
    'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
    'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
    'symmetry_mean', 'texture_worst', 'perimeter_worst', 'area_worst',
    'smoothness_worst', 'compactness_worst', 'concavity_worst',
    'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst'
]

# Set up the plot grid
plt.figure(figsize=(20, 25))

# Loop through each selected feature and create a box plot
for i, feature in enumerate(selected_features, 1):
    # Create a subplot for the current feature
    plt.subplot(6, 3, i)

    # Create a box plot comparing the distribution of the feature for each diagnosis
    sns.boxplot(data=BreastCancer_data, x='diagnosis', y=feature, palette='viridis')

    # Set the title for the current subplot
    plt.title(f'Box Plot of {feature} by Diagnosis', fontsize=12, color='black', weight='bold')
    plt.xlabel('Diagnosis', fontsize=11, color='black', weight='bold')
    plt.ylabel(feature, fontsize=11, color='black', weight='bold')

    # Adjust layout to prevent overlapping titles and labels
    plt.tight_layout()

# Add a main title for all subplots
plt.suptitle('Box Plots of Selected Breast Cancer Features by Diagnosis', fontsize=20, color='black', weight='bold', y=1.02)

# Display the plots
plt.show()
```

**Observations:**
* The box plots clearly illustrate that for most features, malignant tumors show significantly higher median values and wider interquartile ranges compared to benign tumors, indicating larger and more varied cell characteristics.
* Features such as area_worst, perimeter_worst, and radius_worst exhibit a substantial separation between the two diagnosis groups, suggesting they are strong discriminators for tumor type.
* The presence of numerous outliers, particularly in the malignant tumor box plots for several features, highlights the heterogeneity within this group and the importance of considering the full distribution, not just central tendencies.

**6.  How are the various features correlated with each other and with the diagnosis?**

```python
# Calculate the correlation matrix to understand the relationships between numerical features
correlation_matrix = BreastCancer_data.corr(numeric_only=True)

# Generate a heatmap of the correlation matrix
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Breast Cancer Features", fontsize=20, weight='bold')
plt.show()
```

**Observations:**

* The correlation heatmap reveals strong positive correlations among many of the mean, standard error, and "worst" features, particularly within their respective groups (e.g., radius_mean, perimeter_mean, and area_mean are highly correlated).
* Several features, including radius_mean, perimeter_mean, area_mean, concave points_mean, radius_worst, perimeter_worst, and area_worst, show high positive correlations with the diagnosis (when considering a numerical representation of the diagnosis), suggesting they are important predictors of malignancy.
* Features within the 'se' group (standard error) generally show weaker correlations with each other and with the diagnosis compared to the 'mean' and 'worst' features.

**7.  What are the shapes and ranges of the distributions for key features when separated by diagnosis?**

```python
# Select numerical features for distribution plotting (excluding 'id' and 'Unnamed: 32' which were dropped)
numerical_features = BreastCancer_data.select_dtypes(include=np.number).columns

# Set up the plot grid
n_features = len(numerical_features)
n_cols = 4
n_rows = (n_features + n_cols - 1) // n_cols # Calculate number of rows needed

plt.figure(figsize=(n_cols * 5, n_rows * 4))

# Loop through each numerical feature and create a histogram or density plot
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.histplot(data=BreastCancer_data, x=feature, hue='diagnosis', kde=True, palette={'M':'lightblue', 'B':'magenta'})
    plt.title(f'Distribution of {feature} by Diagnosis')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Display the plots
plt.show()
```

**Observations:**

* For many features, there is a noticeable separation in the distributions between benign (B) and malignant (M) tumors, with malignant tumors generally having higher values and wider ranges.
* Features like radius_mean, perimeter_mean, area_mean, concave points_mean, and their 'worst' counterparts show distinct peaks for each diagnosis, indicating their strong potential as discriminators.
* Some features exhibit skewed distributions, and the degree of skewness can differ between benign and malignant cases, providing further insights into the characteristics of each tumor type.

**8.  How do pairs of features relate to each other, and does the relationship differ based on the diagnosis?**

```python
# Select a subset of features for pairplot visualization
# Choosing features that showed good separation in box/violin plots and high correlation
selected_features_for_pairplot = [
    'radius_mean', 'perimeter_mean', 'area_mean',
    'concave points_mean', 'radius_worst', 'perimeter_worst',
    'area_worst', 'concave points_worst', 'diagnosis'
]

# Create a pairplot for the selected features, colored by diagnosis
sns.pairplot(BreastCancer_data[selected_features_for_pairplot], hue='diagnosis', palette={'M':'lightblue', 'B':'magenta'})

# Display the plot
plt.suptitle('Pairwise Relationships of Selected Features by Diagnosis', y=1.02)
plt.show()
```

**Observations:**

* The pairplot reveals clear separation between benign (B) and malignant (M) tumors for many pairs of features, with malignant tumors generally exhibiting higher values and distinct clustering patterns.
* Features within the 'mean' and 'worst' groups show strong positive linear relationships, and the slopes and spreads of these relationships often differ noticeably between the two diagnosis groups.
* The visualizations of individual feature distributions on the diagonal confirm the earlier findings from the histograms and box plots, showing differing shapes and ranges for benign and malignant tumors.

**9.  How do the correlations between pairs of features differ between malignant and benign tumors?**

```python
# Calculate correlation matrix for malignant tumors
# This shows how features relate to each other within the malignant group
malignant_corr = BreastCancer_data[BreastCancer_data['diagnosis']=='M'].drop('diagnosis', axis=1).corr(numeric_only=True)

# Calculate correlation matrix for benign tumors
# This shows how features relate to each other within the benign group
benign_corr = BreastCancer_data[BreastCancer_data['diagnosis']=='B'].drop('diagnosis', axis=1).corr(numeric_only=True)

# Calculate the difference in correlation matrices
# This highlights which feature relationships change the most between malignant and benign tumors
correlation_difference = malignant_corr - benign_corr

# Plot the difference in correlation matrices
plt.figure(figsize=(15, 6))

# Heatmap showing the difference in correlation values
plt.subplot(121)
sns.heatmap(correlation_difference, cmap='RdBu', center=0, vmin=-1, vmax=1, annot=False) # Changed cmap to 'coolwarm'
plt.title('Malignant vs Benign Correlation Differences', color='black', weight='bold')

# Heatmap showing the absolute magnitude of the difference
# This helps identify which correlations have the largest change regardless of direction
plt.subplot(122)
sns.heatmap(correlation_difference.abs(), cmap='viridis', annot=False) # Changed cmap to 'viridis'
plt.title('Absolute Difference Magnitude', color='black', weight='bold')

plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show() # Display the plots
```

**Observations:**

* The differential correlation heatmaps highlight which feature relationships change most significantly between malignant and benign tumors.
* Some pairs of features show stronger correlations in malignant tumors compared to benign tumors, and vice versa, indicating that the interplay between characteristics differs depending on the tumor type.
* The absolute difference heatmap clearly points to the pairs of features where the strength of the relationship varies most substantially between the two diagnosis groups, which can be crucial for feature selection.

**10.  Findings and Concluding Remarks**

**Key Findings and Significant Implications for the development of a Predictive Model**

***Based on the Exploratory Data Analysis of the Breast Cancer Diagnosis dataset, some of the key findings and implications for building a predictive model are as follows:***

*   The dataset exhibits class imbalance, with significantly more benign cases than malignant ones, which will require consideration during model training (e.g., using techniques like stratification, resampling, or adjusting class weights).
*   Many features show distinct distributions and significantly different mean and median values between benign and malignant tumors, particularly features related to the size and irregularity of cell nuclei (radius, perimeter, area, concave points, and their 'worst' counterparts). These features are likely to be strong predictors of diagnosis.
*   High correlations exist among many features, especially within the 'mean' and 'worst' groups, suggesting potential multicollinearity. This might necessitate feature selection or dimensionality reduction techniques to improve model stability and interpretability.
*   The differential correlation analysis revealed that the relationships between some feature pairs change significantly between benign and malignant tumors, providing further discriminatory information that a model could leverage.
*   The presence of outliers, especially in malignant cases for certain features, indicates data variability and might require robust preprocessing steps or models less sensitive to extreme values.
*   The clear separation observed in pairwise plots for several feature combinations reinforces their potential as powerful discriminators for classifying tumor type.
*   Overall, the EDA suggests that a predictive model should focus on features related to cell morphology and nuclear characteristics, while also addressing class imbalance and potential multicollinearity to achieve accurate and reliable breast cancer diagnosis prediction.
